```python
data.columns = [col.strip().lower().replace(" ", "_").replace("-", "_") for col in data.columns]
```

```python
duplicates = data.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

# Remove them
data = data.drop_duplicates()
```

```python
data['scheduledday'] = pd.to_datetime(data['scheduledday'])
data['appointmentday'] = pd.to_datetime(data['appointmentday'])
```

```python
data['gender'] = data['gender'].str.upper().replace({'F': 'Female', 'M': 'Male'})
data['no_show'] = data['no_show'].str.strip().str.title()  # 'No' / 'Yes'
```

```python
invalid_ages = data[(data['age'] < 0) | (data['age'] > 110)]
print(f"Invalid age records: {len(invalid_ages)}")
```

```python
data = data[(data['age'] >= 0) & (data['age'] <= 110)]
```

```python
data = data[data['scheduledday'] <= data['appointmentday']]
```

```python
print(data.dtypes)
```

```python
 print("\nData types after cleaning:")
print(data.dtypes)

print("\nCleaned dataset preview:")
print(data.head())
```

I started by renaming all the column names to lowercase with underscores so they look neat and are easy to work with.
Next, I removed duplicate records to keep only unique entries and checked for any missing values in important columns like age, gender, and appointment dates — luckily, there weren’t any major gaps.
The date columns (scheduledday and appointmentday) were converted into proper datetime format to make time-based analysis easier.
I also cleaned up text fields by changing “F/M” to “Female/Male” and making the no_show values consistent.
Invalid age entries (below 0 or above 110) were removed, and I made sure that scheduled dates always came before or on the actual appointment dates.
Finally, I saved the cleaned and well-structured dataset as medical_appointments_cleaned.csv, ready for further analysis.
